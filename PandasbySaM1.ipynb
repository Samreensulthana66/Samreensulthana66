### Cleaning Data

<p>1)Removing F,C etc strings from Temperature columns.<br>
2)Replace spatial values like -99999, n.a. and No event with NaN.<br>
3)Filling empty cells with NaN values.
</p>

```python
new_df1=df.replace({
    "temp" : "[A-Za-z]",
    "ws"   : "[A-Za-z]"}," ", regex=True)
```

```python
new_df1
```

```python
new_df2=new_df1.replace({
    "temp" :-99999,
    "ws"   :-99999,
    "event":["No event","n.a."] },np.NaN)
```

```python
new_df2
```

```python
new_df3=new_df2.replace(" ",np.NaN)
new_df3
```

## Filling Missing Values

<p>1)Filling Missing temp values with Zer0<br>
2)Filling Missing ws with nextday values using bfill<br>
3)Filling Missing ws with previous day values using ffill.<br>
4)Filling NaN values with linearly interpolated values using interpolate.
</p>

```python
new_df4=new_df3.fillna(0)
```

```python
new_df4
```

```python
new_df5=new_df3.fillna(method="bfill")
new_df5
```

```python
new_df6=new_df3.fillna(method="ffill")
new_df6
```

```python
new_df3
```

```python
new_df7=new_df3.interpolate()
new_df7
```
